```python
ps.sqldf("""

SELECT D.*

FROM df as D
WHERE D.age BETWEEN 5 AND 10

""")
```

```python
ps.sqldf("""

SELECT D.*

FROM df as D

WHERE age<1 AND Sex = 'female'

""")
```

```python
ps.sqldf("""

SELECT D.*

FROM df AS D

WHERE (age<1 AND Sex='female')

OR 

(age>70 AND PClass="1")

""")
```

```python
ps.sqldf("""

SELECT D.*

FROM df AS D

WHERE (D.age<1 AND D.Sex='female')

OR 

(D.age>70 AND D.PClass="1")

""")
```

```python
ps.sqldf("""

SELECT D.*

FROM df as D

WHERE D.Name LIKE "%Mrs.%"

""")
```

```python
ps.sqldf("""

SELECT D.age,
      D.age*D.age AS Age_Sqr,
      D.age+1 AS Age_1
FROM df as D
LIMIT 5

""" )
```

```python
ps.sqldf("""

SELECT D.Sex,
      D.Name,
      D.Sex || D.Name AS Sex_Plus_Name
FROM df AS D

LIMIT 5

""")
```

```python
ps.sqldf("""

SELECT D.Age,
      D.Name,
      D.PassengerId,
      CASE
          WHEN D.age IS NULL THEN 'Missing'
          WHEN D.age >18 THEN 'OLDER'
          WHEN D.age <18 THEN 'YOUNGER'
      ELSE "NO_IDEA" END AS Age_bucket     
FROM df As D

""")
```

```python
ps.sqldf("""

SELECT D.Age,
      D.Name,
      D.PassengerId,
      CASE
          WHEN D.age IS NULL THEN 'Missing'
          WHEN D.age >18 THEN 'OLDER'
          WHEN D.age <18 THEN 'YOUNGER'
      ELSE "NO_IDEA" END AS Age_bucket     
FROM df As D
WHERE Age_bucket = "YOUNGER"
""")
```

```python
ps.sqldf("""

SELECT 
D.Name,

SUBSTR(D.Name,2,4) AS Out

FROM df AS D

""")
```

```python
ps.sqldf("""

SELECT D.Name,

LOWER (D.Name) AS LCase,
UPPER (D.Name) AS UCase

FROM df AS D
""")
```
